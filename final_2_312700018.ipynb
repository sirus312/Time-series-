```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pykalman import KalmanFilter
```

```python
#2.(a)(10%): Estimate the local trend model in Equations (11.1) and (11.2) in the slide Week 8-2.
#Write down the exact equation of your estimated model.
#或用最大概似法
# 读取数据集
data = pd.read_csv('C:/Users/user/OneDrive/文件/時間序列/Dataset-Stock.txt', header=None, names=['volatility'])
# 定义卡尔曼滤波器
# 本地趋势模型方程式:
# yt = µt + et
# µt+1 = µt + ηt
# 其中 et ~ N(0, σ^2_e), ηt ~ N(0, σ^2_η)

# 初始化卡尔曼滤波器参数
initial_state_mean = data['volatility'][0]
kf = KalmanFilter(
    transition_matrices=[1],
    observation_matrices=[1],
    initial_state_mean=initial_state_mean,
    initial_state_covariance=1,
    observation_covariance=1,
    transition_covariance=0.1
)

# 应用卡尔曼滤波器
state_means, state_covariances = kf.filter(data['volatility'].values)

# 绘制原始数据和滤波后的数据
plt.figure(figsize=(8, 4))
plt.plot(data['volatility'], label='Original Data')
plt.plot(state_means, label='Kalman Filtered', color='red')
plt.title('Local Trend Model with Kalman Filter')
plt.xlabel('Time')
plt.ylabel('Volatility')
plt.legend()
plt.show()

# 打印卡尔曼滤波器估计的本地趋势参数
print("Estimated Local Trend (µt):")
print(state_means.flatten())

# 打印原始数据 (yt)
print("Observed Values (yt):")
print(data['volatility'].values)

# 计算和打印残差的方差 (et的方差)
residuals = data['volatility'].values - state_means.flatten()
variance_et = np.var(residuals)
print("Variance of et (Residual Variance):", variance_et)

# 计算和打印状态转移的方差 (ηt的方差)
state_diffs = np.diff(state_means.flatten())
variance_eta = np.var(state_diffs)
print("Variance of ηt (State Transition Variance):", variance_eta)

# 保存 µt 和 yt 的估计值到文件中
results = pd.DataFrame({
    'yt': data['volatility'],
    'mu_t': state_means.flatten()
})
results.to_csv('mu_t_and_yt_estimates.csv', index=False)
```

```python
#2.(b)
# 计算95%置信区间
state_std = np.sqrt(state_covariances[:, 0, 0])
confidence_interval_upper = state_means.flatten() + 1.96 * state_std
confidence_interval_lower = state_means.flatten() - 1.96 * state_std

# 绘制滤波变量及其95%置信区间
plt.figure(figsize=(8, 4))
plt.plot(data.index, data['volatility'], label='Original Data')
plt.plot(data.index, state_means.flatten(), label='Filtered Mean (µt)', color='red')
plt.fill_between(data.index, confidence_interval_lower, confidence_interval_upper, color='pink', alpha=0.5, label='95% Confidence Interval')
plt.title('Filtered Variables with 95% Confidence Interval')
plt.xlabel('Time')
plt.ylabel('Volatility')
plt.legend()
plt.show()
```

```python
#2.(c)
data = pd.read_csv('C:/Users/user/OneDrive/文件/時間序列/Dataset-Stock.txt', header=None, names=['volatility'])

# 定义卡尔曼滤波器
initial_state_mean = data['volatility'][0]
kf = KalmanFilter(
    transition_matrices=[1],
    observation_matrices=[1],
    initial_state_mean=initial_state_mean,
    initial_state_covariance=1,
    observation_covariance=2.9672902745284437,
    transition_covariance=0.4174770598095103
)

# 应用卡尔曼滤波器
state_means, state_covariances = kf.filter(data['volatility'].values)

# 应用卡尔曼平滑器
smoothed_state_means, smoothed_state_covariances = kf.smooth(data['volatility'].values)

# 计算95%置信区间
smoothed_state_std = np.sqrt(smoothed_state_covariances[:, 0, 0])
confidence_interval_upper = smoothed_state_means.flatten() + 1.96 * smoothed_state_std
confidence_interval_lower = smoothed_state_means.flatten() - 1.96 * smoothed_state_std

# 绘制平滑变量及其95%置信区间
plt.figure(figsize=(8, 4))
plt.plot(data.index, data['volatility'], label='Original Data')
plt.plot(data.index, smoothed_state_means.flatten(), label='Smoothed Mean (μt)', color='red')
plt.fill_between(data.index, confidence_interval_lower, confidence_interval_upper, color='pink', alpha=0.5, label='95% Confidence Interval')
plt.title('Smoothed Variables with 95% Confidence Interval')
plt.xlabel('Time')
plt.ylabel('Volatility')
plt.legend()
plt.show()
```
