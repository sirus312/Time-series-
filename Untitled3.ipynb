```python
import statsmodels.api as sm

# 定義參數
n = 200
r = np.zeros(n)
a = np.zeros(n)
sigma2 = np.zeros(n)

# 初始化變數
r[0] = 0
r[1] = 0
a[0] = np.sqrt(-2 * np.log(1 - np.random.random())) * np.cos(2 * np.pi * np.random.random())
sigma2[0] = 0.1

# 生成資料點
for t in range(1, n):
    a[t] = np.sqrt(sigma2[t-1]) * np.sqrt(-2 * np.log(1 - np.random.random())) * np.cos(2 * np.pi * np.random.random())
    sigma2[t] = 0.1 + 0.12 * a[t-1]**2 + 0.09 * sigma2[t-1]
    r[t] = 0.1 + 0.3 * r[t-1] - 0.3 * r[t-2] + a[t]

# 計算 r_t 的自相關函數（ACF）
acf_r = sm.tsa.acf(r, nlags=n-1)

# 計算 r_t^2 的自相關函數（ACF）
acf_r_squared = sm.tsa.acf(r**2, nlags=n-1)

# 計算 |r_t| 的自相關函數（ACF）
acf_abs_r = sm.tsa.acf(np.abs(r), nlags=n-1)

# 繪製 ACF 圖表
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.stem(acf_r)
plt.title('$ r_t $ 的 ACF')
plt.xlabel('滯後')
plt.ylabel('自相關')

plt.subplot(1, 3, 2)
plt.stem(acf_r_squared)
plt.title('$ r_t^2 $ 的 ACF')
plt.xlabel('滯後')
plt.ylabel('自相關')

plt.subplot(1, 3, 3)
plt.stem(acf_abs_r)
plt.title('$ |r_t| $ 的 ACF')
plt.xlabel('滯後')
plt.ylabel('自相關')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm

# 模擬資料
np.random.seed(123)
n = 200
r = np.zeros(n)
a = np.random.normal(0, 1, n)
sigma2 = np.zeros(n)
r[0] = 0
r[1] = 0
sigma2[0] = 0.1

# 生成資料點
for t in range(n):
    if t >= 2:
        sigma2[t-1] = 0.1 + 0.12 * a[t-2]**2 + 0.09 * sigma2[t-2]
    if t >= 1:
        r[t] = 0.1 + 0.3 * r[t-1] - 0.3 * r[t-2] + a[t-1]

# 估計最佳的 ARMA(p,q) 模型
best_aic = np.inf 
best_order = None
best_model = None
for p in range(3):
    for q in range(3):
        try:
            model = sm.tsa.ARMA(r, order=(p, q))
            results = model.fit(method='mle', trend='nc')
            aic = results.aic
            if aic < best_aic:
                best_aic = aic
                best_order = (p, q)
                best_model = results
        except:
            continue

print("Best ARMA(p,q) model:", best_order)
print("AIC:", best_aic)
print(best_model.summary())
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# 模擬資料
np.random.seed(123)
n = 200
r = np.zeros(n)
a = np.random.normal(0, 1, n)
sigma2 = np.zeros(n)
r[0] = 0
r[1] = 0
sigma2[0] = 0.1

# 生成資料點
for t in range(n):
    if t >= 2:
        sigma2[t-1] = 0.1 + 0.12 * a[t-2]**2 + 0.09 * sigma2[t-2]
    if t >= 1:
        r[t] = 0.1 + 0.3 * r[t-1] - 0.3 * r[t-2] + a[t-1]

# 初始化 AIC 表格
results = []
for p in range(3):
    for q in range(3):
        try:
            model = sm.tsa.ARMA(r, order=(p, q))
            results.append((p, q, model.fit(method='mle', trend='nc').aic))
        except:
            continue

# 將 AIC 值轉換成 DataFrame
df = pd.DataFrame(results, columns=['p', 'q', 'AIC'])

# 繪製 AIC 表格
pivot_table = pd.pivot_table(df, values='AIC', index='p', columns='q')
pivot_table = pivot_table.style.background_gradient(cmap='viridis', axis=None)
pivot_table
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# 模擬資料
np.random.seed(123)
n = 200
r = np.zeros(n)
a = np.random.normal(0, 1, n)
sigma2 = np.zeros(n)
r[0] = 0
r[1] = 0
sigma2[0] = 0.1

# 生成資料點
for t in range(n):
    if t >= 2:
        sigma2[t-1] = 0.1 + 0.12 * a[t-2]**2 + 0.09 * sigma2[t-2]
    if t >= 1:
        r[t] = 0.1 + 0.3 * r[t-1] - 0.3 * r[t-2] + a[t-1]

# 初始化 AIC 表格
results = []
for p in range(3):
    for q in range(3):
        try:
            model = sm.tsa.ARIMA(r, order=(p, 0, q))
            fit_model = model.fit(method='mle')
            aic_value = fit_model.aic
            results.append((p, q, aic_value))
        except Exception as e:
            print(f"Failed to fit ARMA({p},{q}): {e}")

# 如果結果列表不為空，則繪製 AIC 表格
if results:
    df = pd.DataFrame(results, columns=['p', 'q', 'AIC'])
    pivot_table = pd.pivot_table(df, values='AIC', index='p', columns='q')
    pivot_table = pivot_table.style.background_gradient(cmap='viridis', axis=None)
    pivot_table
else:
    print("No valid ARMA models found.")
```

```python
import numpy as np
import statsmodels.api as sm

# 模擬資料
np.random.seed(123)
n = 200
r = np.zeros(n)
a = np.random.normal(0, 1, n)
sigma2 = np.zeros(n)
r[0] = 0
r[1] = 0
sigma2[0] = 0.1

# 生成資料點
for t in range(n):
    if t >= 2:
        sigma2[t-1] = 0.1 + 0.12 * a[t-2]**2 + 0.09 * sigma2[t-2]
    if t >= 1:
        r[t] = 0.1 + 0.3 * r[t-1] - 0.3 * r[t-2] + a[t-1]

# 定義 EACF 函數
def eacf(data, max_p, max_q):
    eacf_matrix = np.zeros((max_p+1, max_q+1))
    for p in range(max_p+1):
        for q in range(max_q+1):
            try:
                model = sm.tsa.ARIMA(data, order=(p, 0, q))
                fit_model = model.fit(method='mle', trend='nc')
                eacf_matrix[p, q] = fit_model.bic
            except:
                eacf_matrix[p, q] = np.nan
    return eacf_matrix

# 計算 EACF
max_p = 5
max_q = 5
eacf_matrix = eacf(r, max_p, max_q)

# 打印 EACF 矩陣
print("EACF Matrix:")
print(eacf_matrix)
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# Parameters
n = 200
r = np.zeros(n)
a = np.zeros(n)
sigma_squared = np.zeros(n)

# Initial values
r[0] = 0
r[1] = 0
sigma_squared[0] = 0.1

# Generating data
for t in range(1, n):
    a[t] = np.random.normal(0, 1)
    sigma_squared[t] = 0.1 + 0.12 * a[t-1]**2 + 0.09 * sigma_squared[t-1]
    r[t] = 0.1 + 0.3 * r[t-1] - 0.3 * r[t-2] + np.sqrt(sigma_squared[t]) * a[t]

# Calculate AIC for different ARMA models
results = []
for p in range(4):
    for q in range(4):
        try:
            model = sm.tsa.ARIMA(r, order=(p, 0, q))
            result = model.fit()
            results.append((p, q, result.aic))
        except:
            continue

# Convert results to DataFrame
results_df = pd.DataFrame(results, columns=['P', 'Q', 'AIC'])

# Print the AIC table
print(results_df)
```
