```python
from pydlm import dlm, trend
data = pd.read_csv('C:/Users/user/OneDrive///Dataset-Stock.txt', header=None, names=['volatility'])

# 
myDLM = dlm(data['volatility'])
myDLM = myDLM + trend(degree=1, discount=0.95, name='local trend')

# 
myDLM.fitForwardFilter()

# 
myDLM.plot()

# 
mu = myDLM.getMean(filterType='forwardFilter')

# 
print("Estimated Local Trend Model Parameters:")
print("mu_t = ", mu[:-1])
print("eta_t = ", np.diff(mu))

# 
print("Last Local Trend Value (mu_{T}):", mu[-1])
```
