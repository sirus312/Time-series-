```python
#1
import numpy as np

# Simulating X(t)
np.random.seed(42)  # For reproducibility
X1 = np.random.normal(0, 1, 100)
X2 = np.random.normal(1, 1, 100)

# Concatenating the two datasets
X = np.concatenate((X1, X2))
X
```

```python
#2
from scipy.stats import norm
import matplotlib.pyplot as plt

# Fitting X with a single normal distribution
mu, std = norm.fit(X)

# Plotting histogram and fitted distribution
plt.hist(X, bins=30, density=True, alpha=0.6, color='g')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
plt.title("Fit result: mu = %.2f,  std = %.2f" % (mu, std))
plt.show()
```

```python
#3
from scipy.stats import norm
from sklearn.mixture import GaussianMixture

# Fitting X with a mixture of normals
gmm = GaussianMixture(n_components=2, random_state=42)
gmm.fit(X.reshape(-1, 1))

# Plotting histogram and fitted mixture of normals
plt.hist(X, bins=30, density=True, alpha=0.6, color='g')
x = np.linspace(xmin, xmax, 1000)
pdf = np.exp(gmm.score_samples(x.reshape(-1, 1)))
plt.plot(x, pdf, color='blue', alpha=0.6, label='GMM')
plt.title('Fitting with Mixture of Normals')
plt.legend()
plt.show()
```

```python
#4
import os
os.environ["OMP_NUM_THREADS"] = "1"
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import GridSearchCV

# Finding the optimal number of components
params = {'n_components': np.arange(1, 6)}
gmm = GaussianMixture(random_state=42)
grid = GridSearchCV(gmm, params)
grid.fit(X.reshape(-1, 1))

best_gmm = grid.best_estimator_

# Plotting histogram and fitted mixture of normals with best number of components
plt.hist(X, bins=30, density=True, alpha=0.6, color='g')
pdf = np.exp(best_gmm.score_samples(x.reshape(-1, 1)))
plt.plot(x, pdf, color='blue', alpha=0.6, label='GMM')
plt.title('Fitting with Optimal Number of Components')
plt.legend()
plt.show()
```
