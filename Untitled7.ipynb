```python
import numpy as np
import pandas as pd

# Load dataset
data = pd.read_csv('C:/Users/user/OneDrive/文件/時間序列/Data-GEstock.txt', delimiter="\t", header=None)
returns = data[0].values
returns
```

```python
# Initialize parameters
np.random.seed(42)
iterations = 10000
beta1 = np.random.randn()
beta2 = np.random.randn()
alpha10 = np.random.randn()
alpha11 = np.random.randn()
alpha12 = np.random.randn()
alpha20 = np.random.randn()
alpha21 = np.random.randn()
alpha22 = np.random.randn()
e1 = np.random.rand()
e2 = np.random.rand()

# Initialize latent state variables
T = len(returns)
s = np.random.choice([1, 2], size=T)

# Initialize conditional variances
h = np.zeros(T)
h[0] = np.var(returns)

# Gibbs sampling iterations
for _ in range(iterations):
    # Sample h_t based on the current state s_t
    for t in range(1, T):
        if s[t] == 1:
            h[t] = alpha10 + alpha11 * h[t-1] + alpha12 * returns[t-1]**2
        else:
            h[t] = alpha20 + alpha21 * h[t-1] + alpha22 * returns[t-1]**2

    # Sample the state s_t
    for t in range(T):
        p1 = (e1 * np.exp(-0.5 * ((returns[t] - beta1 * np.sqrt(h[t]))**2 / h[t]))) / np.sqrt(2 * np.pi * h[t])
        p2 = (e2 * np.exp(-0.5 * ((returns[t] - beta2 * np.sqrt(h[t]))**2 / h[t]))) / np.sqrt(2 * np.pi * h[t])
        s[t] = 1 if np.random.rand() < p1 / (p1 + p2) else 2

    # Sample beta1 and beta2 based on the current states and returns
    beta1 = np.random.normal(np.mean(returns[s == 1] / np.sqrt(h[s == 1])), np.std(returns[s == 1] / np.sqrt(h[s == 1])))
    beta2 = np.random.normal(np.mean(returns[s == 2] / np.sqrt(h[s == 2])), np.std(returns[s == 2] / np.sqrt(h[s == 2])))

    # Sample transition probabilities e1 and e2
    e1 = np.random.beta(1 + np.sum(s[:-1] == 1), 1 + np.sum(s[:-1] == 2))
    e2 = np.random.beta(1 + np.sum(s[:-1] == 2), 1 + np.sum(s[:-1] == 1))

# Final estimated parameters
print(f'beta1: {beta1}, beta2: {beta2}')
print(f'alpha10: {alpha10}, alpha11: {alpha11}, alpha12: {alpha12}')
print(f'alpha20: {alpha20}, alpha21: {alpha21}, alpha22: {alpha22}')
print(f'e1: {e1}, e2: {e2}')
```
